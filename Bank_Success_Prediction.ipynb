Perform the following activities using Bank Marketing dataset.

1. Consider duration, age, and campaign columns as
predictors and the column “y”(states whether the client will subscribe a term deposit or not) as target  variable

2. Use 70% of the data as training data set and 30% of data as testing data set.

3. Build a logistic regression model.

4. Determine the classification accuracy rate of the model on both train and test set.

```python

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
df = pd.read_csv('/content/drive/My Drive/Infosys ML Course/bank-additional-full.csv')
df.head()
```

```python
data= df[['age','campaign', 'y']]
data.head()
```

```python
data.shape
```

```python
x = data[['age','campaign']]
y= data['y']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.3, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
train_pred = model.predict(x_train)
test_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

cm_train = confusion_matrix(y_train, train_pred)
cm_test = confusion_matrix(y_test, test_pred)

pd.DataFrame(cm_train, columns=model.classes_, index = model.classes_)
# cm_test
```

```python
pd.DataFrame(cm_test, columns=model.classes_, index = model.classes_)
#
```

```python
correct_pred = cm_train[0,0] + cm_train[1,1]
total_pred = cm_train.sum()
accuracy = correct_pred/total_pred
accuracy
```

```python
accuracy = model.score(x_train, y_train)
accuracy
```
